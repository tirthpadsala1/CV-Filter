```python
from langchain_community.document_loaders import PyPDFLoader
from utils.var import content

loader = PyPDFLoader(content["exCV"])
CV = loader.load()
CV
```

```python
text = CV[0].page_content
text
```

```python
import re
text = text.lower()
PunctuationRemoved_text = re.sub(r"[^a-z0-9\s]", " ", text)
PunctuationRemoved_text
```

```python
metadata = CV[0].metadata
metadata
```

```python
len(PunctuationRemoved_text)
```

```python
import json

wJson = json.loads(metadata['wozbermeta'])
values = (
    wJson
        .get("templateStyle", {}) 
        .get("headers", {})
        .get("text", {})
        .get("values", {})
)
```

```python
values
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

textSplitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50,
    separators=["\n\n", "\n", ".", " ", ""]
)

spittedText = textSplitter.split_text(PunctuationRemoved_text)
spittedText[:1000]
```

```python
from sentence_transformers import SentenceTransformer

EmbeddingModel = SentenceTransformer("BAAI/bge-base-en-v1.5")

vectors = EmbeddingModel.encode(
    spittedText,
    batch_size = 16,
    normlize_embeddings=True,
    show_progress_bar=True
)


```
