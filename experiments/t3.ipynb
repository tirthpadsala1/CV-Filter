```python
from utils.var import content
import pdfplumber
import sys

def extract_text(path:str):

    try:
        text = "" 

        with pdfplumber.open(path) as f:
            for pageNum , page in enumerate(f.pages):
                pageText = page.extract_text()
                if pageText:
                    text += f"--( page:{pageNum +1} )--\n {pageText}"
        return text
    except Exception as e:
        print(e)
    



textSy = extract_text(content['exPDF'])
```

```python
import os
import re

def check_file_name(path:str):
    
    try:

        fileName = os.path.basename(path).lower()

        for pattern in filenames:
            if re.match(fileName , pattern):
                return True,f"filename matches pattern:{pattern}"
            
    
        return False,"filename dosen't match any pattern"



    except Exception as e:
        print(e)
```

```python

def keywordChecker2(text: str, keywords):

    text = text.lower()

    # remove punctuation
    text = re.sub(r"[^a-z0-9\s]", " ", text)

    words = text.split()

    # create bigrams
    bigrams = [f"{words[i]} {words[i+1]}" for i in range(len(words)-1)]

    score = 0
    matched = set()


    for w,bg in zip_longest(words , bigrams , fillvalue="xx"):
        if w in keywords["high"] and w not in matched:
            score += 3
            matched.add(w)
        if w in keywords["medium"] and w not in matched:
            score += 1
            matched.add(w)

        if bg in keywords["high"] and bg not in matched:
            score += 3
            matched.add(bg)

        if bg in keywords["medium"] and bg not in matched:
            score += 1
            matched.add(bg)

    if score >= 15:
        return True, f"confirmed CV, score={score}"

    elif score >= 8:
        return True, f"might be CV, score={score}"

    else:
        return False, f"not a CV, score={score}"
```
